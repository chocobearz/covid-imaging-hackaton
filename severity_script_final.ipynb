```python
import pandas as pd
df1 = pd.read_csv("metadata.csv")
df2 = pd.read_csv("covid-severity-scores.csv")
```

```python
df2
```

```python
result = pd.merge(df1, df2, how='right', on=['filename'])
result
```

```python
result['Severity_score'] = result['geographic_mean'] + result['opacity_mean']
result
```

```python
result1=result.dropna(subset = ["offset"])
```

```python

result1.astype({'offset': 'int32'}).dtypes
```

```python
result2=result1.dropna(subset=["intubated"])
result2.shape
```

```python
#sum((result2['went_icu']=="N"))
result3=result2.dropna(subset=['went_icu','age'])
result3 
```

```python
result3.went_icu[result3.went_icu == 'Y'] = 0.0
result3.went_icu[result3.went_icu== 'N'] = 1.0
result3.intubated[result3.intubated == 'Y']= 0.0
result3.intubated[result3.intubated == 'N']= 1.0
result3.sex[result3.sex == 'M']= 0.0
result3.sex[result3.sex == 'F']= 1.0

```

```python
result3
```

```python
X = result3[['offset', 'age', 'sex', 'went_icu', 'intubated']].values
y = result3['Severity_score'].values
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm

X_train
```

```python
#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1 = df.head(5)
df1
```

```python
import numpy as np
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
